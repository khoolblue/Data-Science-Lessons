```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)

df = pd.read_csv("data_cleaning_jupyter.csv")
#to display top 5 rows
df.head(5)
```

```python
df.tail(5)
```

```python
df.dtypes
```

```python
#dropping/removing unwanted columns
df.drop(["Engine Fuel Type", "Market Category", "Vehicle Style", "Popularity", "Number of Doors", "Vehicle Size"], axis = 1)
df.head(5)
```

```python
#Renaming the columns
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
df_duplicated = df[df.duplicated()]
print("Number of duplicated cells are: ",df_duplicated.shape)
```

```python
df.count
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.shape
```

```python
print(df.isnull().sum())
```

```python
#bar chart 
df.Make.value_counts().nlargest(40).plot(kind = "bar", figsize = (10,5))
plt.title("Number of Cars by titles")
plt.ylabel("Number of Cars")
plt.xlabel("Make")
plt.show()
```

```python
#Scatter plot
fig, ax = plt.subplots(figsize = (10,6))
ax.scatter(df["HP"], df["Price"])
ax.set_xlabel("HP")
ax.set_ylabel("Price")
plt.show()
```
